```python
# Standardize the folder:
# - remove spaces in between the name and replace them with single underscore
# - label the folder as healthy in case it has no label after double underscore (__)
# - replace the hyphen with underscore
unlabeled_folder = ['Black-grass', 'Charlock', 'Cleavers', 'Common Chickweed', 'Common wheat', 'Fat Hen', 'Loose Silky-bent', 'Maize',
                    'Scentless Mayweed', 'Shepherd’s Purse', 'Small-flowered Cranesbill', 'Sugar beet']
len(unlabeled_folder)
# the plants with only healthy label, help us to classify (identify) the plant -- (at least)
```

```python
unlabeled_folder = ['Black-grass', 'Charlock', 'Cleavers', 'Common Chickweed', 'Common wheat', 'Fat Hen', 'Loose Silky-bent', 'Maize',
                    'Scentless Mayweed', 'Shepherd’s Purse', 'Small-flowered Cranesbill', 'Sugar beet']

def process_name(name):
    name = name.replace(' ', '_')
    name = name.replace('-', '_')
    name = name + '__healthy'
    return name

processed_list = [process_name(name) for name in unlabeled_folder]

for item in processed_list:
    print(item)
```

```python
name_mapping = {original: processed for original, processed in zip(unlabeled_folder, processed_list)}

name_mapping
```

```python
# Rename the filder_name column elements in the dataframe
df['folder_name'] = df['folder_name'].map(name_mapping).fillna(df['folder_name'])
```

```python
# Background_without_leaves a folder in dataset:
# As this folder is random pictures of healthy plants and pictures which are not plants at all (random things image). So, I labeled it as unknown.
# It does not lie in any class (neither in plants nor disease labels)
# Consideration:
# Anything or plants having no disease label can be considered as healthy.
# This approach will not be harmfull as label it as otherwise (any disease label).
# It will help us in differentiating between a leaf (plants) and a random background
# Later we tune our model with different paramaters in order to get a well trained model to meet our need, then we could find a meaning for it.
# df.rename(columns={'Background_without_leaves': 'Background_without_leaves__healthy'})
```

```python
pd.DataFrame(df.folder_name.unique())
```

```python
# add two more columns extracting from folder name which are plant and disease
def split_folder_name(folder_name):
    if '__' in folder_name:
        plant, disease = folder_name.split('__', 1)
        disease = disease.lstrip('_')
        return plant, disease
    else:
        # print('value of folder_name, unknown', folder_name, 'Unknown')
        return folder_name, 'Unknown'


df[['plant', 'disease']] = df['folder_name'].apply(lambda x: pd.Series(split_folder_name(x)))
```

```python
df[['plant', 'disease']]
```

```python
# Assume the folder_name as target variable
# change the folder_name to target
df.rename(columns = {'folder_name': 'target'}, inplace=True)
```

**create a binary columns for healthy and unhealthy plants**

```python
# create a binary column for healthy and unhealthy plants
df['is_healthy'] = df['disease'].apply(lambda x: 1 if x == 'healthy' else 0)
```

```python
# normalize the numerical value
from sklearn.preprocessing import StandardScaler

numerical_cols = ['width_px', 'height_px', 'chn_0_px_avg', 'chn_1_px_avg', 'chn_2_px_avg']

# Initializing the StandardScaler
scaler = StandardScaler()

# Standardizing the numerical columns
normalized_data = filtered_data.copy()
normalized_data[numerical_cols] = scaler.fit_transform(normalized_data[numerical_cols])

# Plotting the pairplot
sns.pairplot(normalized_data[['width_px', 'height_px', 'chn_0_px_avg', 'chn_1_px_avg', 'chn_2_px_avg', 'plant', 'disease']], hue='plant')
plt.show()
```

```python
# normalize the numerical value with MinMaxScalar
from sklearn.preprocessing import MinMaxScaler

numerical_cols = ['width_px', 'height_px', 'chn_0_px_avg', 'chn_1_px_avg', 'chn_2_px_avg']

min_max_scalar = MinMaxScaler()

normalized_data1 = filtered_data.copy()
normalized_data1[numerical_cols] = min_max_scalar.fit_transform(normalized_data1[numerical_cols])

sns.pairplot(normalized_data1[['width_px', 'height_px', 'chn_0_px_avg', 'chn_1_px_avg', 'chn_2_px_avg', 'plant', 'disease']], hue='plant')
plt.show()
```
